```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import io
df= pd.read_excel(io.BytesIO(uploaded['Test_set.xlsx']))
```

```python
df.head()
```

```python
import io
Data_df= pd.read_excel(io.BytesIO(uploaded['Data_Train.xlsx']))
```

```python
Data_df.head()
```

```python
##Merge two dataset

final_df = pd.concat([Data_df, df], ignore_index = True)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
##Feature Engineering Method
final_df["Day"] = final_df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
final_df["Month"] = final_df['Date_of_Journey'].apply(lambda x:x.split("/")[1])
final_df["Year"] = final_df['Date_of_Journey'].apply(lambda x:x.split("/")[2])
```

```python
final_df.head()
```

```python
## Data types change

final_df['Day'] = final_df['Day'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
##Drop the colm date of jrny

final_df.drop('Date_of_Journey', axis= 1, inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(1)
```

```python
## arrival time split bcz time and moth boths are together

final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
##check missing value

final_df.isnull().sum()
```

```python
## seprate hour and min from arrival time

final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_min'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df.head(1)
```

```python
##change data types

final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
##drop the clm

final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
##Again seprate hour&time from Deptime

final_df['dep_hour'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['dep_min'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[1])
```

```python
##change data type
final_df['dep_hour'] = final_df['dep_hour'].astype(int)
final_df['dep_min'] = final_df['dep_min'].astype(int)
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
## replace with number

final_df['Total_Stops']=final_df['Total_Stops'].replace({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':'nan', '4 stops':4})
```

```python
final_df.head(5)
```

```python
## drop route column

final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Additional_Info'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
